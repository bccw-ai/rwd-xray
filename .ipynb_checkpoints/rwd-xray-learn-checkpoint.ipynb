```python
# -*- coding: utf-8 -*-
import sys
import struct
import operator
import itertools
import re             # 正则表达式是独立于任何语言的一种字符串匹配表达式，
```

```python
def _get_decoder(key1, key2, key3, op1, op2, op3):
    decoder = {}
    values = set()

    for e in range(256):
        d = op3(op2(op1(e, key1), key2), key3) & 0xFF
        decoder[chr(e)] = chr(d)  # 返回Unicode码 为整数 i 的字符的 字符串格式
        values.add(d)

    return decoder if len(values) == 256 else None
```
